```python
import pandas as pd
import numpy as np

import bs4
from urllib.request import urlopen
from bs4 import BeautifulSoup

from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def pdfdownload(url,filename):
    import urllib.request
    pdf_path =url #"http://distcourts.tap.nic.in/hcorders/2014/arbappl/arbappl_132_2014.pdf"
    print(url)
    
    
    response = urllib.request.urlopen(url)    
    file = open(r"H:/court/2016/pdffiles/may/"+filename+".pdf", 'wb')
    file.write(response.read())
    file.close()
```

```python
def Main():
    driver=webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    driver.get("http://ghconline.nic.in/JDMTMay16.html")

    sleep(2)

   
    #driver.find_element_by_id("searchtwofour").click()

    sleep(5)
    page_source = driver.page_source
    soup1 = BeautifulSoup(page_source, 'lxml')
    type(soup1)
    bs4.BeautifulSoup
    title = soup1.title


    PDF_file_Table_tag=soup1.find_all("table",{"class":"DISCOVERY3"})[0]

    tr_tag=PDF_file_Table_tag.find_all("tr")
    print(len(tr_tag))
    
    Date_of_Judgment=[]
    CASE_NO=[]
    PARTIES_NAME=[]
    SUBJECT=[]
    JUDGMENT_DELIVERED_BY=[]

    for i in range(2,len(tr_tag)):
        td_tag1=tr_tag[i].find_all("td")[0]
        td_tag2=tr_tag[i].find_all("td")[1]
        td_tag3=tr_tag[i].find_all("td")[2]
        td_tag4=tr_tag[i].find_all("td")[3]
        td_tag5=tr_tag[i].find_all("td")[4]
        td_tag6=tr_tag[i].find_all("td")[5]
        
        
        Date_of_Judgment.append(td_tag2.text)
        CASE_NO.append(td_tag3.text)
        PARTIES_NAME.append(td_tag4.text)
        SUBJECT.append(td_tag5.text)
        JUDGMENT_DELIVERED_BY.append(td_tag6.text)
        

    

        td_tag=tr_tag[i].find_all("td")[2]
        
        document=td_tag3.text
        print(document)
        
        
        
        try:
            a_tag_url=td_tag.find("a")["href"]
            tag="http://ghconline.nic.in/"+a_tag_url.replace("./","")
            pdfdownload(tag,document.replace("/"," "))
            
        
        except:
            print("No Document")
            
    df = pd.DataFrame(data={"Date_of_Judgment":Date_of_Judgment,"CASE_NUMBER": CASE_NO,"PARTIES_NAME":PARTIES_NAME,"SUBJECT": SUBJECT,"JUDGMENT_DELIVERED_BY":JUDGMENT_DELIVERED_BY})
    df.to_csv(r"H:\court\2016\csvfiles\may.csv", sep=',',index=False)
    df.head(50)

```
