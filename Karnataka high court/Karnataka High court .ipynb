```python
import pandas as pd
import numpy as np


import bs4
from urllib.request import urlopen
from bs4 import BeautifulSoup


from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def pdfdownload(url,filename):
    import urllib.request
    pdf_path =url
    print(url)
    
    
    response = urllib.request.urlopen(url)    
    file = open(r"H:/court/2016/pdffiles/ma/"+filename+".pdf", 'wb')
    file.write(response.read())
    file.close()
```

```python
def Main():
    driver=webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    driver.get("http://judgmenthck.kar.nic.in/judgments/browse?type=datecreated&order=ASC&rpp=65&value=2012-01-03")

    sleep(2)

   
    #driver.find_element_by_id("searchtwofour").click()

    sleep(5)
    page_source = driver.page_source
    soup1 = BeautifulSoup(page_source, 'lxml')
    type(soup1)
    bs4.BeautifulSoup
    title = soup1.title


    PDF_file_Table_tag=soup1.find_all("table",{"class":"miscTable"})[0]

    tr_tag=PDF_file_Table_tag.find_all("tr")
    print(len(tr_tag))
    
    Case_Number=[]
    Judge=[]
    Petitioner=[]
    Respondent=[]
    Bench=[]
    for j in range(1,5):

        for i in range(2,len(tr_tag)):
            td_tag1=tr_tag[i].find_all("td")[0]
            td_tag2=tr_tag[i].find_all("td")[1]
            td_tag3=tr_tag[i].find_all("td")[2]
            td_tag4=tr_tag[i].find_all("td")[3]
            td_tag5=tr_tag[i].find_all("td")[4]

            #driver.find_element_by_xpath("/html/body/table[2]/tbody/tr[1]/td[2]/table[2]/tbody/tr[2]/td[1]").click()
            print(td_tag1)
            a_tag_url=td_tag1.find("a")["href"]
            print(a_tag_url)
            driver.get("http://judgmenthck.kar.nic.in/"+a_tag_url)

            page_source = driver.page_source
            soup2 = BeautifulSoup(page_source, 'lxml')
            type(soup2)
            bs4.BeautifulSoup
            title = soup2.title
            td_tag=soup2.find("td",{"headers":"t1"})
            a_tag_url=td_tag.find("a")["href"]

            document=td_tag1.text
            print(document)

            tag="http://judgmenthck.kar.nic.in/"+a_tag_url.replace("./","")
            pdfdownload(tag,document.replace("/"," "))           




            Case_Number.append(td_tag1.text)
            Judge.append(td_tag2.text)
            Petitioner.append(td_tag3.text)
            Respondent.append(td_tag4.text)
            Bench.append(td_tag5.text)

            driver.back()
        try:
            driver.find_element_by_xpath("/html/body/table[2]/tbody/tr[1]/td[2]/div[6]/a").click()
        except NoSuchElementExeception :
            break
    

        
            
    df = pd.DataFrame(data={"Case_Number":Case_Number,"Judge":Judge,"Petitioner":Petitioner,"Respondent":Respondent," Bench": Bench})
    df.to_csv(r"H:\court\2016\csvfiles\maya.csv", sep=',',index=False)
    df.head(50)

```
